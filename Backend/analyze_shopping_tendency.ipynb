```python
# load csv file
import pandas as pd

df = pd.read_csv('./MigrosData/Migros_case/Shoppin_Cart/trx_202309.csv')
df["ArtikelID"] = df["ArtikelID"].astype(str)
```

```python
from app.database.database import Database
database = Database()
```

```python
df = df.head(3430)
```

```python
# store average sustainability rank for each customer
df["SustainabilityRank"] = df["ArtikelID"].apply(database.get_product_sustainability_rank)
df.groupby("KundeID")["SustainabilityRank"].mean().sort_values(ascending=False)
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(df.groupby("KundeID")["SustainabilityRank"].mean().sort_values(ascending=False).head(77))
```

```python
df_vis = pd.DataFrame()
df_vis["avg"] = df.groupby("KundeID")["SustainabilityRank"].mean().sort_values(ascending=False).head(77)
```

```python
df_vis["index"] = df_vis.index
```
